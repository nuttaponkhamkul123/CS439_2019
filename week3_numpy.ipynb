# Masking


```python

import numpy as np
```

```python
x = np.arange(0,6*np.pi,0.1)
x.shape
```

```python
y = np.sin(x)+np.random.randn(x.shape[0])*0.5
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('ggplot')

plt.plot(x,y,'.r')
```

```python
y[y<0] = 0
plt.plot(x,y,'.g')
```
