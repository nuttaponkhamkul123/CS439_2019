```python
import pandas as pd
```

```python
df_member = pd.read_csv('member.csv')
df_room = pd.read_csv('room.csv')
df_rent = pd.read_csv('rent.csv')
df_member.drop(labels=['Unnamed: 0','mobile'],axis=1,inplace=True)
df_room.drop(labels=['Unnamed: 0','isavailable'],axis=1,inplace=True)
df_rent.drop(labels=['Unnamed: 0'],axis=1,inplace=True)
df_member.head()
```

```python
df = df_rent\
    .merge(df_member,left_on='member_id',right_on='id')\
    .drop(labels=['id_y'],axis=1)\
    .merge(df_room,left_on='room_id',right_on='id')\
    .drop(labels=['id'],axis=1)
df.head()
```

# Q1 ใครจ่ายเงินค่าเช่ามากที่สุด

```python
df1 = df.groupby(by=['member_id','name'])\
    .agg({'cost' : ['sum']})\
    .sort_values(('cost','sum'),ascending=False)\
    .head(1)
df1
```

# Q2 ห้องเช่าใดทำรายได้มากที่สุด

```python
df2 = df.groupby(by=['room_id','number'])\
    .agg({'cost' : ['sum']})\
    .sort_values(('cost','sum'),ascending=False)\
    
df2
```

# Q7 หาชื่อสมาชิกที่เช่าห้อง 8 ชั่วโมงในเดือนมิถุนายน

```python
from datetime import datetime, timedelta
init = datetime.strptime( "2018-06-01T00:00:00", "%Y-%m-%dT%H:%M:%S" )
final = datetime.strptime( "2018-06-30T23:59:59", "%Y-%m-%dT%H:%M:%S" )
df7 = df

df7.stop = pd.to_datetime(df7['stop'],format='%Y-%m-%d %H:%M:%S')
df7.start = pd.to_datetime(df7['start'],format='%Y-%m-%d %H:%M:%S')
df7=df[(init<=df.stop) & (df.stop<=final)]

df7['hours'] = df7.stop-df7.start
df7 = df7.groupby('name').agg({'hours':'sum'})
df8 = df7
df7.loc[df7.hours=='08:00:00']




# df7
```
