```python
import pandas as pd
```

```python
df_member = pd.read_csv('member.csv')
df_room = pd.read_csv('room.csv')
df_rent = pd.read_csv('rent.csv')
df_member.drop(labels=['Unnamed: 0','mobile'],axis=1,inplace=True)
df_room.drop(labels=['Unnamed: 0','isavailable'],axis=1,inplace=True)
df_rent.drop(labels=['Unnamed: 0'],axis=1,inplace=True)
df_member.head()
```

```python
df_room.head()
```

```python
df_rent.head()
```

```python
df_rent.count()
```

```python
df_room.count()
```

```python
df = df_rent\
    .merge(df_member,left_on='member_id',right_on='id')\
    .drop(labels=['id_y'],axis=1)\
    .merge(df_room,left_on='room_id',right_on='id')\
    .drop(labels=['id'],axis=1)
df.head()
```

# Q1 ใครจ่ายเงินค่าเช่ามากที่สุด

```python
df1 = df.groupby(by=['member_id','name'])\
    .agg({'cost' : ['sum']})\
    .sort_values(('cost','sum'),ascending=False)\
    .head(1)
df1
```

# Q2 ห้องเช่าใดทำรายได้มากที่สุด

```python
df2 = df.groupby(by=['room_id','number'])\
    .agg({'cost' : ['sum']})\
    .sort_values(('cost','sum'),ascending=False)\
    
df2
```

# Q5 ในเดือนมิถุนายนมีรายรับรวมเท่าใด

```python
from pytz import datetime
df5 = df.filter(items=['start','stop','cost'],axis=1)
time_start = df5.start
time_stop = df5.stop
# delta_time = time_stop - time_stop

to_date_start = pd.to_datetime(time_start,format='%Y-%m-%d %H:%M:%S')
to_date_stop = pd.to_datetime(time_stop,format='%Y-%m-%d %H:%M:%S')

df5
```

```python
# df5['delta'] = to_date_stop - to_date_start
df5['new_date_start'] = to_date_start
df
df5['start_month'] = df5.new_date_start.apply(lambda x : x.month)
df5['stop_month'] = df5.new_date_stop.apply(lambda x : x.month)
df5
```
